````python
# GPA Classification System

# 1. Get input
gpa = float(input("Enter student's GPA (0.0 - 4.0): "))

# 2. Validate GPA range
if gpa < 0.0 or gpa > 4.0:
    print("Invalid GPA. Please enter a value between 0.0 and 4.0.")

# 3. Classification logic (highest condition first)
elif gpa >= 3.5:
    print("Status: Dean's List")

elif gpa >= 3.0:
    print("Status: Good Standing")

elif gpa >= 2.0:
    print("Status: Probation")

else:
    print("Status: Failing")
````

````python
# Study Recommendation System

# 1) Get inputs
while True:
    try:
        sleep_hours = float(input("How many hours did you sleep last night? "))
        if sleep_hours < 0:
            print("Invalid input: sleep hours cannot be negative. Please try again.")
            continue
        break
    except ValueError:
        print("Invalid input: Please enter a number for sleep hours.")

while True:
    try:
        stress_level = int(input("What is your stress level (1-10)? "))
        if stress_level < 1 or stress_level > 10:
            print("Invalid input: stress level must be between 1 and 10. Please try again.")
            continue
        break
    except ValueError:
        print("Invalid input: Please enter an integer between 1 and 10 for stress level.")

while True:
    try:
        days_until_exam = int(input("How many days until your next exam? (Enter a number) "))
        if days_until_exam < 0:
            print("Invalid input: days until exam cannot be negative. Please try again.")
            continue
        break
    except ValueError:
        print("Invalid input: Please enter an integer for days until exam.")

# 2) Validate inputs (simple checks - moved into input loops for robustness)

# 3) Decision logic
if sleep_hours < 6:
    print("Recommendation: Get more rest. Study in short, focused sessions (e.g., 25 minutes study + 5 minutes break).")
elif stress_level >= 7:
    print("Recommendation: Reduce workload today. Take breaks and use stress management (walk, breathing, or light exercise).")
else:
    # Optional: tailor the 'else' based on exam proximity
    if days_until_exam <= 2:
        print("Recommendation: Exam is soon. Do focused revision and practice questions with a clear plan.")
    elif days_until_exam <= 7:
        print("Recommendation: Build a daily plan. Review notes and do practice problems each day.")
    else:
        print("Recommendation: Start early. Study steadily and review regularly (spaced practice).")
````

````python
# Study Recommendation System (based on study hours, stress level, and exam timing)

studied_hours = float(input("How many hours did you study today? "))
stress_level = int(input("What is your stress level (1-10)? "))
exam_input = input("When is your next exam? (e.g., 'tomorrow', 'in 3 days', '7') ").strip().lower()

# Convert exam_input to days_until_exam
if "tomorrow" in exam_input:
    days_until_exam = 1
elif "day" in exam_input:
    # e.g., "in 3 days" or "3 days"
    tokens = exam_input.replace("in", "").strip().split()
    days_until_exam = int(tokens[0])
else:
    # assume it's a number like "7"
    days_until_exam = int(exam_input)

# Basic validation
if studied_hours < 0:
    print("Invalid input: studied hours cannot be negative.")
elif stress_level < 1 or stress_level > 10:
    print("Invalid input: stress level must be between 1 and 10.")
elif days_until_exam < 0:
    print("Invalid input: days until exam cannot be negative.")
else:
    # Decision logic (priority: exam urgency -> stress -> low study time -> else)
    if days_until_exam <= 1:
        print("Recommendation: Exam is very soon. Do a focused review + practice questions. Avoid distractions.")
    elif stress_level >= 7:
        print("Recommendation: Stress is high. Use shorter study blocks with breaks (e.g., 25/5) and reduce workload.")
    elif studied_hours < 2:
        print("Recommendation: You studied very little. Do a short, clear session (30–60 mins) to build momentum.")
    else:
        print("Recommendation: Keep a steady plan. Review notes + do practice problems consistently.")
````
