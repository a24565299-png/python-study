## Part 1: Enhanced Expense Entry (15 minutes)

### 🔄 From One to Many

Transform your Week 2 single-expense tracker to handle multiple expenses with proper validation:

```python
print("=== Enhanced Finance Tracker v0.3 ===")
print("Week 3: Loop-Powered Edition")
print()

# User setup (enhanced from Week 2)
name = input("Enter your name: ")

# Add budget validation using while loop
monthly_budget = 0
while monthly_budget < 100 or monthly_budget > 10000:
    try:
        budget_input = input("Enter monthly budget ($100-$10,000): $")
        monthly_budget = float(budget_input)

        if monthly_budget < 100:
            print("Budget too low! Minimum $100")
        elif monthly_budget > 10000:
            print("Budget seems high! Maximum $10,000")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nWelcome {name}! Monthly budget: ${monthly_budget:.2f}")

# ========== MULTI-EXPENSE ENTRY ==========
expenses = []          # Store all expense amounts
descriptions = []      # Store what each expense was for
total_spent = 0

print("\n--- Enter Your Expenses ---")
print("Type 'done' when finished entering expenses")

# Multi-expense entry system
while True:
    expense_input = input("\nExpense amount (or 'done'): $")

    if expense_input.lower() == 'done':
        break

    try:
        expense_amount = float(expense_input)

        if expense_amount <= 0:
            print("Expense must be greater than $0!")
            continue

        description = input("What was this expense for? ").strip()

        while description == "":
            print("Description cannot be empty!")
            description = input("What was this expense for? ").strip()

        expenses.append(expense_amount)
        descriptions.append(description)
        total_spent += expense_amount

        budget_percentage = (total_spent / monthly_budget) * 100

        print(f"✅ Added: {description} - ${expense_amount:.2f}")
        print(f"💰 Running total: ${total_spent:.2f}")
        print(f"📈 Budget used: {budget_percentage:.1f}%")

        if total_spent > monthly_budget:
            print("⚠️ Warning: You are over budget!")

    except ValueError:
        print("Please enter a valid number or 'done'.")

print(f"\n📊 Expense entry complete: {len(expenses)} expenses, ${total_spent:.2f} total")
```

## Part 2: Basic Expense Analysis (15 minutes)

### 📈 Processing Your Transaction Data

Use loops to analyze the expenses you collected:

```python
# Expense Analysis Section
if len(expenses) > 0:
    print("\n📊 Expense Analysis:")

    # Basic statistics using loop patterns from workshops
    print(f"Total expenses: {len(expenses)}")
    print(f"Total spent: ${total_spent:.2f}")
    print(f"Average expense: ${total_spent/len(expenses):.2f}")
    print(f"Budget used: {(total_spent/monthly_budget)*100:.1f}%")
    print(f"Remaining budget: ${monthly_budget - total_spent:.2f}")

    # Find highest and lowest expenses
    highest_expense = expenses[0]
    lowest_expense = expenses[0]

    for expense in expenses:
        if expense > highest_expense:
            highest_expense = expense
        if expense < lowest_expense:
            lowest_expense = expense

    print(f"\n💰 Expense Range:")
    print(f"Highest expense: ${highest_expense:.2f}")
    print(f"Lowest expense: ${lowest_expense:.2f}")

    # Categorize expenses by size
    # Large (>20% of budget), Medium (5-20%), Small (<5%)
    large_count = 0
    medium_count = 0
    small_count = 0

    for expense in expenses:
        percentage = (expense / monthly_budget) * 100

        if percentage > 20:
            large_count += 1
        elif percentage >= 5:
            medium_count += 1
        else:
            small_count += 1

    print(f"\n📊 Expense Categories:")
    print(f"Large expenses (>20% budget): {large_count}")
    print(f"Medium expenses (5-20% budget): {medium_count}")
    print(f"Small expenses (<5% budget): {small_count}")

    # Smart feedback based on analysis
    print("\n💡 Smart Feedback:")

    budget_used = (total_spent / monthly_budget) * 100

    if budget_used > 100:
        print("⚠️ You are over budget. You should reduce unnecessary spending.")
    elif budget_used >= 80:
        print("⚠️ You have used most of your budget. Spend carefully for the rest of the month.")
    elif budget_used >= 50:
        print("✅ Your spending is moderate, but keep monitoring your expenses.")
    else:
        print("✅ Great job! Your spending is well under control.")

    if large_count >= 2:
        print("💸 You have several large expenses. Review them carefully.")
    elif large_count == 1:
        print("📌 One large expense had a big impact on your budget.")
    else:
        print("👍 Most of your expenses are small or medium.")

else:
    print("\n📝 No expenses entered - try adding some transactions!")
```

## Part 3: Multiple Savings Goals Tracker (15 minutes)

### 🎯 Goal Management with Loops

Extend your tracker to handle multiple savings goals:

```python
print("\n=== Savings Goals Tracker ===")

# Goal collection
goal_names = []
goal_targets = []
goal_current = []

# Ask how many goals they want to track
num_goals = 0
while num_goals < 1 or num_goals > 5:  # Reasonable limits
    try:
        num_goals = int(input("How many savings goals? (1-5): "))
        if num_goals < 1:
            print("Must have at least 1 goal!")
        elif num_goals > 5:
            print("Maximum 5 goals for this tracker!")
    except ValueError:
        print("Please enter a valid number!")

print(f"\nEntering {num_goals} savings goals:")

for i in range(num_goals):
    print(f"\n--- Goal {i+1} ---")

    # Goal name
    goal_name = input("Goal name: ").strip()
    while goal_name == "":
        print("Goal name cannot be empty!")
        goal_name = input("Goal name: ").strip()

    # Target amount
    target_amount = 0
    while target_amount <= 0:
        try:
            target_amount = float(input("Target amount: $"))
            if target_amount <= 0:
                print("Target must be greater than $0!")
        except ValueError:
            print("Please enter a valid number!")

    # Current savings
    current_amount = -1
    while current_amount < 0 or current_amount > target_amount:
        try:
            current_amount = float(input("Current savings: $"))
            if current_amount < 0:
                print("Current savings cannot be negative!")
            elif current_amount > target_amount:
                print("Current savings cannot exceed the target!")
        except ValueError:
            print("Please enter a valid number!")

    # Store values
    goal_names.append(goal_name)
    goal_targets.append(target_amount)
    goal_current.append(current_amount)

print(f"\n✅ Saved {len(goal_names)} goals!")
```

```python
# Make sure the lists exist (prevents NameError)
goal_names = []
goal_targets = []
goal_current = []

# Goal Analysis and Progress Tracking
if len(goal_names) > 0:
    print("\n📊 Goals Progress Analysis:")

    total_target = 0
    total_current = 0

    lowest_progress = 100
    lowest_goal = ""

    for i in range(len(goal_names)):

        progress = (goal_current[i] / goal_targets[i]) * 100
        remaining = goal_targets[i] - goal_current[i]
        months_needed = remaining / 100 if remaining > 0 else 0

        print(f"\nGoal: {goal_names[i]}")
        print(f"Target: ${goal_targets[i]:.2f}")
        print(f"Saved: ${goal_current[i]:.2f}")
        print(f"Progress: {progress:.1f}%")
        print(f"Remaining: ${remaining:.2f}")
        print(f"Months to completion (saving $100/month): {months_needed:.1f}")

        total_target += goal_targets[i]
        total_current += goal_current[i]

        # Track lowest progress goal
        if progress < lowest_progress:
            lowest_progress = progress
            lowest_goal = goal_names[i]

    # Overall progress summary
    overall_progress = (total_current / total_target) * 100 if total_target > 0 else 0
    total_remaining = total_target - total_current

    print(f"\n🎯 Overall Goals Summary:")
    print(f"Total target: ${total_target:.2f}")
    print(f"Total saved: ${total_current:.2f}")
    print(f"Overall progress: {overall_progress:.1f}%")
    print(f"Total remaining: ${total_remaining:.2f}")

    print("\n⚡ Goal Needing Most Attention:")
    print(f"{lowest_goal} has the lowest progress ({lowest_progress:.1f}%).")
    print("Consider focusing more savings on this goal.")

else:
    print("No goals to analyze!")
```

## Part 4: Simple Menu System (10 minutes)

### 🎮 Interactive Program Flow

Create a basic menu that lets users choose what to do:

```python
print("\n=== Interactive Finance Menu ===")

# Simple menu system
running = True

while running:
    print("\n" + "-"*30)
    print(f"💰 {name}'s Finance Tracker")
    if len(expenses) > 0:
        print(f"📊 Current: {len(expenses)} expenses, ${sum(expenses):.2f} spent")
    print("-"*30)
    print("1. View expense summary")
    print("2. View goals progress")
    print("3. Add new expense")
    print("4. Budget check")
    print("5. Exit")

    choice = input("\nChoose option (1-5): ")

    if choice == "1":
        # Display expense summary
        print("\n--- Expense Summary ---")

        if len(expenses) == 0:
            print("No expenses recorded yet.")
        else:
            for i in range(len(expenses)):
                print(f"{i+1}. ${expenses[i]:.2f} - {descriptions[i]}")
            print(f"Total spent: ${sum(expenses):.2f}")

    elif choice == "2":
        # Display goals progress
        print("\n--- Goals Progress ---")

        if len(goal_names) == 0:
            print("No savings goals recorded yet.")
        else:
            for i in range(len(goal_names)):
                progress = (goal_current[i] / goal_targets[i]) * 100
                print(f"{goal_names[i]}: ${goal_current[i]:.2f}/${goal_targets[i]:.2f} ({progress:.1f}% complete)")

    elif choice == "3":
        # Add single new expense
        print("\n--- Add New Expense ---")

        while True:
            try:
                new_expense = float(input("Enter new expense amount: $"))
                if new_expense <= 0:
                    print("Expense must be greater than $0!")
                else:
                    break
            except ValueError:
                print("Please enter a valid number!")

        new_description = input("What was this expense for? ").strip()
        while new_description == "":
            print("Description cannot be empty!")
            new_description = input("What was this expense for? ").strip()

        expenses.append(new_expense)
        descriptions.append(new_description)

        print(f"✅ Added: ${new_expense:.2f} - {new_description}")
        print(f"New total spent: ${sum(expenses):.2f}")

        if sum(expenses) > monthly_budget:
            print("⚠️ Warning: You are now over budget!")

    elif choice == "4":
        # Budget status check
        print("\n--- Budget Check ---")

        total_spent = sum(expenses)
        budget_used = (total_spent / monthly_budget) * 100
        remaining = monthly_budget - total_spent
        daily_budget_left = remaining / 30

        print(f"Monthly budget: ${monthly_budget:.2f}")
        print(f"Total spent: ${total_spent:.2f}")
        print(f"Budget used: {budget_used:.1f}%")
        print(f"Remaining budget: ${remaining:.2f}")
        print(f"Suggested daily budget left: ${daily_budget_left:.2f} per day")

        if budget_used > 100:
            print("⚠️ You are over budget. Cut back on spending.")
        elif budget_used >= 80:
            print("⚠️ You are close to your limit. Spend carefully.")
        elif budget_used >= 50:
            print("✅ You are using your budget at a moderate pace.")
        else:
            print("✅ Your budget is in a healthy range.")

    elif choice == "5":
        # Exit with summary
        running = False
        print(f"\nGoodbye {name}!")
        print(f"Final summary: {len(expenses)} expenses, ${sum(expenses) if expenses else 0:.2f} total")
        print("🚀 Your finance tracker is growing stronger each week!")

    else:
        print("Invalid choice! Please choose 1-5.")

print("\n=== Finance Tracker Session Complete ===")
```
