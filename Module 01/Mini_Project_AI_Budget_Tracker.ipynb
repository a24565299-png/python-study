### Paste and Test Your AI-Generated Budget Tracker

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
Write a Python program that acts as a simple monthly budget tracker.
The program should ask the user to input their monthly income.
Then it should ask the user to enter up to three expenses.
After collecting the income and expenses, the program should calculate the remaining budget.
Finally, it should display the remaining budget and show a warning message if the remaining budget is negative.
"""

# Ask for monthly income
income = float(input("Enter your monthly income: "))

# Ask for up to 3 expenses
expense1 = float(input("Enter expense 1: "))
expense2 = float(input("Enter expense 2: "))
expense3 = float(input("Enter expense 3: "))

# Calculate total expenses
total_expenses = expense1 + expense2 + expense3

# Calculate remaining budget
remaining_budget = income - total_expenses

# Display results
print("\n----- Budget Summary -----")
print(f"Monthly Income: ${income}")
print(f"Total Expenses: ${total_expenses}")
print(f"Remaining Budget: ${remaining_budget}")

# Warning if budget is negative
if remaining_budget < 0:
    print("⚠ Warning: Your expenses exceed your income!")
else:
    print("✅ Your budget is within your income.")
```
