# Part 2: Guided Practice with Sample Data (30 minutes)

Let's practice AI collaboration with real business data before moving to your assignment.

## 📁 Loading Sample Data

First, let's load our sample transaction data:

````python
import pandas as pd
import os

# Step 1: Check current files
print("📁 Current files in directory:")
print(os.listdir())

# Step 2: If transactions.csv doesn't exist, create sample data
if 'transactions.csv' not in os.listdir():
    print("\n⚠️ transactions.csv not found. Creating sample file...")

    data = {
        "Date": ["2026-03-01", "2026-03-02", "2026-03-03", "2026-03-04"],
        "Amount": ["$12.50", "$45.00", "$20.00", "$100.00"],
        "Category": ["Food", "Transport", "Food", "Shopping"],
        "Description": ["Lunch", "Taxi", "Dinner", "Shoes"]
    }

    df_sample = pd.DataFrame(data)
    df_sample.to_csv("transactions.csv", index=False)

    print("✅ Sample transactions.csv created!")

# Step 3: Load the data
df = pd.read_csv('transactions.csv')

# Step 4: Display basic info
print("\n📋 Sample Transaction Data:")
print("=" * 40)
print(df.head())

print(f"\n📊 Dataset Info:")
print(f"Total transactions: {len(df)}")
print(f"Date range: {df['Date'].min()} to {df['Date'].max()}")
print(f"Categories: {df['Category'].unique()}")

# Step 5: Clean Amount column
df['Amount'] = pd.to_numeric(
    df['Amount'].str.replace('$', '', regex=False).str.replace(',', '', regex=False),
    errors='coerce'
)

# Remove invalid rows
df = df.dropna(subset=['Amount'])

# Step 6: Calculate category spending
category_spending = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)

# Step 7: Display results
print("\n💰 Spending Summary by Category")
print("=" * 40)
for category, amount in category_spending.items():
    print(f"{category:<15}: ${amount:>8.2f}")

# Step 8: Total spending
total_spending = category_spending.sum()
print(f"\n📊 Total Spending: ${total_spending:.2f}")

# Step 9: Highest spending category
top_category = category_spending.idxmax()
top_amount = category_spending.max()
print(f"🏆 Highest Spending Category: {top_category} (${top_amount:.2f})")

# Step 10: Budget check
budget = 150
if total_spending > budget:
    print("⚠️ Warning: You are over your budget.")
else:
    print("✅ Good job: You are within your budget.")
````

## 🤖 AI Collaboration Exercise 1: Basic Analysis

**Your Turn**: Practice prompting AI for basic data analysis.

## Practice Prompt Template

Try asking AI:
*"Using this transaction DataFrame, I want to understand my spending patterns. Please create code that shows:

Format the output to be easy to read for a business presentation."*

**Expected AI Response** (try this yourself first!):

````python
# Step: Basic Spending Analysis

# Group by Category and calculate metrics
spending_analysis = df.groupby('Category')['Amount'].agg(
    Total_Spent='sum',
    Avg_Amount='mean',
    Transaction_Count='count'
).round(2)

# Sort by total spending
spending_analysis = spending_analysis.sort_values('Total_Spent', ascending=False)

# Display results
print("💼 Business Spending Analysis")
print("=" * 50)
print(f"{'Category':<15} {'Total':<10} {'Average':<10} {'Count':<8}")
print("-" * 50)

for category, row in spending_analysis.iterrows():
    total = row['Total_Spent']
    avg = row['Avg_Amount']
    count = row['Transaction_Count']

    print(f"{category:<15} ${total:<9.2f} ${avg:<9.2f} {count:<8.0f}")

# Grand total
print(f"\nGrand Total: ${spending_analysis['Total_Spent'].sum():.2f}")
````

## 📈 Step 2: Create Core Analysis Functions

**Prompt AI** to help you create the foundation of your finance assistant:

::: {.callout-tip}
## Analysis Functions Prompt:

*"For my Smart Finance Assistant, I need functions to analyze spending data. Please create well-documented functions that:
1. Calculate total spending by category
2. Find the top 5 spending categories
3. Calculate average spending per transaction
4. Identify any unusual transactions (very high or negative amounts)
Each function should have clear business explanations and error handling."*
:::

````python
# AI-generated analysis functions
# (Your AI collaboration code goes here)

# Core Analysis Functions for Smart Finance Assistant

# Core Analysis Functions for Smart Finance Assistant

def calculate_total_spending_by_category(df):
    """
    Calculate total spending for each category.

    Business purpose:
    Helps identify where most money is being spent.
    """
    try:
        if df.empty:
            return "Error: The dataset is empty."

        result = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
        return result

    except KeyError as e:
        return f"Error: Missing required column: {e}"
    except Exception as e:
        return f"Unexpected error: {e}"


def get_top_5_spending_categories(df):
    """
    Return the top 5 categories with the highest total spending.

    Business purpose:
    Helps focus on the biggest expense areas first.
    """
    try:
        if df.empty:
            return "Error: The dataset is empty."

        result = df.groupby('Category')['Amount'].sum().sort_values(ascending=False).head(5)
        return result

    except KeyError as e:
        return f"Error: Missing required column: {e}"
    except Exception as e:
        return f"Unexpected error: {e}"


def calculate_average_spending_per_transaction(df):
    """
    Calculate the average transaction amount.

    Business purpose:
    Shows the typical amount spent per purchase.
    """
    try:
        if df.empty:
            return "Error: The dataset is empty."

        average = df['Amount'].mean()
        return round(average, 2)

    except KeyError as e:
        return f"Error: Missing required column: {e}"
    except Exception as e:
        return f"Unexpected error: {e}"


def identify_unusual_transactions(df, high_threshold=100):
    """
    Identify unusual transactions such as very high amounts or negative values.

    Business purpose:
    Helps detect abnormal spending, possible errors, or refunds.
    """
    try:
        if df.empty:
            return "Error: The dataset is empty."

        unusual = df[(df['Amount'] > high_threshold) | (df['Amount'] < 0)]
        return unusual

    except KeyError as e:
        return f"Error: Missing required column: {e}"
    except Exception as e:
        return f"Unexpected error: {e}"
````

## 💰 Step 3: Generate Business Insights

**Challenge AI** to help you create business-relevant insights:

::: {.callout-tip}
## Business Insights Prompt:

*"Using the transaction analysis functions, help me create a comprehensive spending report that would be useful for personal financial planning. Include recommendations for potential savings and spending pattern observations. Format it like a professional financial summary."*
:::

````python
# AI-generated business insights
# (Your AI collaboration code goes here)

# Business Insights Report

def generate_business_insights(df):
    try:
        # Use previous functions
        total_by_category = calculate_total_spending_by_category(df)
        top_5 = get_top_5_spending_categories(df)
        avg_spending = calculate_average_spending_per_transaction(df)
        unusual = identify_unusual_transactions(df, high_threshold=100)

        total_spending = total_by_category.sum()
        top_category = total_by_category.idxmax()
        top_amount = total_by_category.max()

        print("📊 PERSONAL FINANCIAL REPORT")
        print("=" * 50)

        print(f"\n💰 Total Spending: ${total_spending:.2f}")
        print(f"🏆 Highest Spending Category: {top_category} (${top_amount:.2f})")
        print(f"📈 Average Transaction: ${avg_spending:.2f}")

        print("\n📂 Top 5 Spending Categories:")
        for category, amount in top_5.items():
            print(f"- {category}: ${amount:.2f}")

        print("\n🔍 Spending Insights:")
        print(f"- You are spending the most on {top_category}, which may be a key area to review.")

        if avg_spending > 50:
            print("- Your average transaction amount is relatively high. Consider reducing large purchases.")
        else:
            print("- Your average spending per transaction is moderate.")

        if len(unusual) > 0:
            print("- There are unusual transactions detected. Review them carefully.")
        else:
            print("- No unusual transactions detected.")

        print("\n💡 Recommendations:")
        print(f"- Focus on reducing expenses in {top_category}.")
        print("- Track your spending regularly to stay within budget.")
        print("- Consider setting category-based spending limits.")

    except Exception as e:
        print(f"Error generating insights: {e}")


# Run the report
generate_business_insights(df)
````
